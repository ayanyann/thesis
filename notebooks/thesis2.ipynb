```python
# import libraries again
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
```

```python
# porters stemmer 
stemmer = PorterStemmer()

# Function to tokenize, remove stopwords, and stem texts 
def process_text(text):
# convert non-string data to string 
    text = str(text)
    tokens = word_tokenize(text)
    # remove stopwords 
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    
    # stem words 
    stemmed = [stemmer.stem(word) for word in tokens]
    
    return''.join(stemmed)

# apply text processing function to each text column
data['Title_Transformed'] = data['Title'].apply(process_text)
data['Keywords_Transformed'] = data['Keywords'].apply(process_text)
data['AbstractNote_Transformed'] = data['Abstract Note'].apply(process_text)

# Concatenate the transformed text into a new column
data['Transformed_Text'] = data['Title_Transformed'] + ' ' + data['Keywords_Transformed'] + ' ' + data['AbstractNote_Transformed']

# Drop the individual transformed columns if you only need the concatenated column
data.drop(['Title_Transformed', 'Keywords_Transformed', 'AbstractNote_Transformed'], axis=1, inplace=True)

# Save the transformed dataset to a new CSV file
data.to_csv('transformed_dataset2.csv', index=False)

# Provide the path for download
print('Transformed dataset saved as "transformed_dataset.csv". Ready to download!')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Combine the text columns into a single string per document
data['Combined_Text'] = data['Transformed_Text']

# Initialize the CountVectorizer
count_vectorizer = CountVectorizer()

# Fit and transform the combined text data
dtm = count_vectorizer.fit_transform(data['Combined_Text'])

# Initialize the LDA model
lda_model = LatentDirichletAllocation(n_components=10, random_state=0)  # You can change the number of topics

# Fit the model to the document-term matrix
lda_model.fit(dtm)

# Retrieve the words per topic
words = count_vectorizer.get_feature_names_out()
topic_summaries = []
for topic_idx, topic in enumerate(lda_model.components_):
    top_features_ind = topic.argsort()[:-26:-1]  # get top 10 keywords for each topic
    top_features = [words[i] for i in top_features_ind]
    topic_summaries.append(" ".join(top_features))

# Display the topics
for i, topic_summary in enumerate(topic_summaries):
    print(f"Topic {i}: {topic_summary}")
    
```

```python
# Initialize the CountVectorizer
count_vectorizer = CountVectorizer()

# Fit and transform the combined text data
dtm = count_vectorizer.fit_transform(data['Combined_Text'])

# Initialize the LDA model
lda_model = LatentDirichletAllocation(n_components=10, random_state=0)  # You can change the number of topics

# Fit the model to the document-term matrix and transform the data
lda_topic_matrix = lda_model.fit_transform(dtm)

# Assign each document to the topic with the highest probability
data['Topic'] = lda_topic_matrix.argmax(axis=1)

# Retrieve the words per topic
words = count_vectorizer.get_feature_names_out()
topics = []
for topic_idx, topic in enumerate(lda_model.components_):
    top_features_ind = topic.argsort()[:-26:-1]  # get top 10 keywords for each topic
    top_features = [words[i] for i in top_features_ind]
    topics.append(" ".join(top_features))

# Display the topics
for i, topic_keywords in enumerate(topics):
    print(f"Topic {i}: {topic_keywords}")

# Group by the assigned topic and publication year, then count the number of documents
topic_counts = data.groupby(['Topic', 'Publication Year']).size().reset_index(name='Num_Documents')

# Display the number of papers per topic and year
print(topic_counts)
```

```python
# TOPIC COHERENCE SCORE

from gensim import corpora
from gensim.models.coherencemodel import CoherenceModel
import nltk

# Tokenize the documents
tokenized_docs = [nltk.word_tokenize(doc.lower()) for doc in data['Combined_Text']]

# Create a Gensim dictionary from the tokenized documents
dictionary = corpora.Dictionary(tokenized_docs)

# Convert the tokenized documents to a corpus
corpus = [dictionary.doc2bow(doc) for doc in tokenized_docs]

# Retrieve the list of terms for each topic
topic_terms = []
for topic_idx, topic in enumerate(lda_model.components_):
    top_features_ind = topic.argsort()[:-11:-1]  # get top 10 keywords for each topic
    topic_terms.append([words[i] for i in top_features_ind])

# Compute Coherence Score using c_v
coherence_model_lda = CoherenceModel(topics=topic_terms, texts=tokenized_docs, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()

print('Coherence Score (c_v):', coherence_lda)
```

# K-Means Cluster

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# Calculate the sum of squared distances for a range of number of clusters
sum_of_squared_distances = []
K_range = range(1, 15)  # You can choose the range of K you want to evaluate

for K in K_range:
    km = KMeans(n_clusters=K, random_state=0)
    km = km.fit(lda_topic_matrix)
    sum_of_squared_distances.append(km.inertia_)

# Plotting the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(K_range, sum_of_squared_distances, 'bx-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Sum of squared distances')
plt.title('Elbow Method For Optimal K')
plt.show()
```

```python
kmeans_model = KMeans(n_clusters=10,random_state=0)
#fiting the model with chosen clusters
kmeans_model.fit(lda_topic_matrix)

# assign cluster per document
cluster_assignments = kmeans_model.labels_
#assign
data['Cluster'] = cluster_assignments


cluster_counts = data['Cluster'].value_counts().sort_index()
print(cluster_counts)
```

```python
# Get the top terms per cluster
print("Top terms per cluster:")
order_centroids = kmeans_model.cluster_centers_.argsort()[:, ::-1]
terms = count_vectorizer.get_feature_names_out()

for i in range(10):  
    print("Cluster %d:" % i, end='')
    for ind in order_centroids[i, :25]:  # You can change 10 to the number of words you want to display
        print(' %s' % terms[ind], end='')
    print()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.cluster import KMeans

# Get the words that are most characteristic of each cluster (centroids)
def get_top_words_for_each_cluster(model, num_words, feature_names):
    top_words = {}
    for idx, topic in enumerate(model.cluster_centers_):
        sorted_words = topic.argsort()[-num_words:]
        top_words[idx] = [feature_names[i] for i in sorted_words]
    return top_words

# 'count_vectorizer' and 'lda_model' are already fitted as from the previous steps
feature_names = count_vectorizer.get_feature_names_out()
top_words = get_top_words_for_each_cluster(kmeans_model, 10, feature_names)

# Show the cluster labels (top words)
for cluster, words in top_words.items():
    print(f"Cluster {cluster}: {words}")

# Now, let's show the papers in each cluster
# We add the cluster assignments to the data DataFrame
data['Cluster'] = kmeans_model.labels_

# Show papers for each cluster
for cluster in sorted(data['Cluster'].unique()):
    print(f"\nCluster {cluster} papers:")
    cluster_papers = data.loc[data['Cluster'] == cluster, 'Title']
    print(cluster_papers.to_string(index=False))
```
